input = output = one-hot binary code 0 to 7
Refer to: https://blog.keras.io/building-autoencoders-in-keras.html

```python
import keras
from keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Define input = output
data = np.arange(0,16,1)
print(data)
# convert the array data into one hot encoding
dataIn = keras.utils.to_categorical(data)
in_size = 16
# This is our input image
# input_img = keras.Input(shape=(8,))
```

```python
# This is the size of our encoded representations
encoding_dim = 4  # 32 floats -> compression of factor 24.5, assuming the input is 784 floats

Inputs = keras.Input(shape=(in_size,)) 
print(Inputs.shape)
# "encoded" is the encoded representation of the input
encoded = layers.Dense(encoding_dim, activation='sigmoid', name = 'Encoder')(Inputs)
# "decoded" is the lossy reconstruction of the input
decoded = layers.Dense(in_size, activation='sigmoid', name = 'Decoder')(encoded)  # decoder layer


# This model maps an input to its reconstruction
autoencoder = keras.Model(Inputs, decoded)
# This model maps inputs to the 4-bit encoder output
encoder = keras.Model(Inputs, encoded)
```

```python
from keras.utils import plot_model
plot_model(autoencoder, to_file='model.png',show_shapes=True, show_layer_names=True)
```

```python
# Now compile the models
autoencoder.compile(optimizer='adam', loss='MeanSquaredError')
# Run unsupervised training on the training set. We run the training for 50 epochs.
autoencoder.fit(dataIn, dataIn, epochs=1000, batch_size=256,shuffle=True, validation_data=(dataIn, dataIn))
```

```python
# Now we want to visualize the encoder activation value and decoder output
out = autoencoder.predict(dataIn)
print(dataIn)
print(out)
# This part to be completed.
```
