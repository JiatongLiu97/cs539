## Cluster Demo

Compare kmeans, spectral, and DBSCAN using 4 datasets:

data_uniform.csv, data_noisycircle.csv, data_noisymoon.csv, data_3blobs.csv

```python
import pandas as pd
import numpy as np
# read the data files.
X = [[]] * 4
X[0] = pd.read_csv('data_uniform.csv',header=None).to_numpy()
X[1] = pd.read_csv('data_noisycircle.csv',header=None).to_numpy()
X[2] = pd.read_csv('data_noisymoon.csv',header=None).to_numpy()
X[3] = pd.read_csv('data_3blobs.csv',header=None).to_numpy()
```

Kmeans algorithm

```python
from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

two_means = cluster.MiniBatchKMeans(n_clusters=2)
for i in range(len(X)):
  two_means.fit(X[i])
  y_pred = two_means.labels_.astype(np.int)
  plt.subplot(1, 4, i+1)
  plt.scatter(X[i][:, 0], X[i][:, 1], color=colors[y_pred].tolist(), s=10)
plt.show()
```

Spectral Clustering

```python
from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

spectral = cluster.SpectralClustering(n_clusters=2,
                                          eigen_solver='arpack',
                                          affinity="nearest_neighbors")
for i in range(len(X)):
  spectral.fit(X[i])
  y_pred = spectral.labels_.astype(np.int)
  plt.subplot(1, 4, i+1)
  plt.scatter(X[i][:, 0], X[i][:, 1], color=colors[y_pred].tolist(), s=10)
plt.show()
```

DBSCAN: Density Based Spatial Clustering of Applications with Noise

```python
from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

dbscan = cluster.DBSCAN(eps=.1,min_samples=5)
for i in range(len(X)):
  dbscan.fit(X[i])
  y_pred = dbscan.labels_.astype(np.int)
  plt.subplot(1, 4, i+1)
  plt.scatter(X[i][:, 0], X[i][:, 1], color=colors[y_pred].tolist(), s=10)

plt.show()
```
